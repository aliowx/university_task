```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from math import sqrt
```

```python
movies_df = pd.read_csv('movies.csv')
movies_df.head()
```

```python
movies_df['year'] = movies_df.title.str.extract('(\(\d\d\d\d\))',expand=False)
movies_df['year'] = movies_df.year.str.extract('(\d\d\d\d)',expand=False)

#Removing the years from the 'title' column
movies_df['title'] = movies_df.title.str.replace('(\(\d\d\d\d\))', '' , regex=True)

movies_df['title'] = movies_df['title'].apply(lambda x: x.strip())
movies_df.head()
```

```python
movies_df['genres'] = movies_df.genres.str.split('|')
movies_df.head()
```

```python
moviesWithGenres_df = movies_df.copy()

for index , row in movies_df.iterrows():
    for genre in row['genres']:
        moviesWithGenres_df.at[index , genre] = 1
moviesWithGenres_df = moviesWithGenres_df.fillna(0)
moviesWithGenres_df.head()
```

```python
ratings_df = pd.read_csv('ratings.csv')
ratings_df.head()
```

```python
ratings_df = ratings_df.drop('timestamp', axis=1)
ratings_df.head()
```

```python
userInput = [
            {'title':'Breakfast Club, The', 'rating':5},
            {'title':'Toy Story', 'rating':3.5},
            {'title':'Jumanji', 'rating':2},
            {'title':"Pulp Fiction", 'rating':5},
            {'title':'Akira', 'rating':4.5}
         ] 
inputMovies = pd.DataFrame(userInput)
inputMovies
```

```python
inputid = movies_df[movies_df['title'].isin(inputMovies['title'])]  
inputid
```

```python
inputMovies = pd.merge(inputid , inputMovies)
inputMovies
```

```python
inputMovies = inputMovies.drop(['genres','year'], axis=1)
inputMovies
```

```python
userMovies = moviesWithGenres_df[ moviesWithGenres_df['movieId'].isin(inputMovies['movieId']) ]
userMovies
```

```python
userMovies = userMovies.reset_index(drop=True)
usergenretable = userMovies.drop(['movieId','title','genres','year'],axis=1)
usergenretable
```

```python
inputMovies['rating']
```

```python
usergenretable.transpose()
```

```python
userprofile = usergenretable.transpose().dot(inputMovies['rating'])
userprofile
```

```python
genreTable = moviesWithGenres_df.set_index(moviesWithGenres_df['movieId'])
genreTable = genreTable.drop(['movieId','title','genres','year'],axis=1)
genreTable.head()
```

```python
genreTable.shape
```

```python
recommendationTable_df = genreTable*userprofile
recommendationTable_df
```

```python
#                               محاسبه مجموع در راستای سطرها
recommendationTable_df = (recommendationTable_df.sum(axis=1))/(userprofile.sum())
recommendationTable_df
```

```python
recommendationTable_df = recommendationTable_df.sort_values(ascending=False)
recommendationTable_df[:10]
```
